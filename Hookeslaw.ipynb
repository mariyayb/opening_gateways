```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_spring(x,color,linestyle,label=None,title=None):
    #points = [[right_x,np.cosh(right_x)],[left_x,np.cosh(left_x)]]

    fig = plt.figure(figsize=(10,25))
    ax = fig.add_subplot(1, 1, 1)

    if not label:
        label = ['Equilibrium' if i == 1 else 'displacement from equilibrium' for i in x]

    for i in range(len(x)):
        
 

        plt.plot(np.linspace(0,1,10),[0]*10,color=color[i]) 

        x_spring = np.linspace(1,1+np.pi*x[i],100)
        plt.plot(x_spring,np.sin((x_spring-1)*12/x[i]),color=color[i],linestyle=linestyle[i]) #the spring

        plt.plot(np.linspace(1+np.pi*x[i],1+np.pi*x[i]+1,10),[0]*10,color=color[i],linestyle=linestyle[i])
        mass_location = 2 + np.pi*x[i]
        #the mass
        plt.plot([mass_location,mass_location+2.5],[-1.25,-1.25],color=color[i],linestyle=linestyle[i]) #bottom
        plt.plot([mass_location,mass_location+2.5],[1.25,1.25],color=color[i],linestyle=linestyle[i]) #top
        plt.plot([mass_location,mass_location],[-1.25,1.25],color=color[i],linestyle=linestyle[i]) #left
        plt.plot([mass_location+2.5,mass_location+2.5],[-1.25,1.25],color=color[i],linestyle=linestyle[i],label=label[i]) #right

    # Move left y-axis and bottim x-axis to centre, passing through (0,0)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position(('axes',0))


    # Eliminate upper and right axes
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    # Show ticks in the left and lower axes only
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    plt.yticks([], [])
    plt.xticks([2 + np.pi*j for j in x],[str(round((np.pi*(j-1)),2)) for j in x])

    plt.xlim((0, 13))
    plt.ylim((-1.25, 3.5))
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(title)
    plt.legend(loc=1)
    num = int(round(x[0],1)*10)
    plt.savefig('static/hookeslaw_{}.png'.format(num),bbox_inches="tight")

    plt.show()

for i in np.arange(0.1,2.5,0.1):
    print(i)
    plot_spring([i,1],color=['teal','gray'],linestyle=['solid','dotted'],label = None )
```
